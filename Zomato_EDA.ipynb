## Working with Second Dataset - Country Code

```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head(100)
```

```python
#Exploring the columns of the two DataFrames
```

```python
df.columns
```

```python

df_country.columns
```

```python
final_df=pd.merge(df,df_country, on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
## To check data types
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
# Maximum number of restaurants are available in India
```

```python
country_values=final_df.Country.value_counts().values
```

```python
## Cretaing a Pie Chart for this Country distribution
# Top 3 countries that use Zomato
plt.pie( x=country_values[:3],labels=country_names[:3], autopct='%0.2f%%');
```

The top three countries where Zomato have transactions are India, US and Brazil respectively in the order of the transactions

```python
final_df.columns
```

```python
final_df.head(3)
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
final_df.groupby(['Rating color','Rating text']).agg(['min', 'max'])
#This is just to see the range of values of each olumn when grouped -  not of much ue anyways
```

```python
ratings.head()
```

Obsevations
1. Whenever the rating is between 4.5 to 4.9 -> Excellent
2. Whenever the ratings are between 4 to 4.4 -> Very Good 
3. Whenever the ratings are between 3.5 to 3.9 -> Good
4. Whenever the ratings are between 3.0 to 3.4 -> Average
  

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot( x='Aggregate rating',y='Rating Count',data=ratings)
# This is a normal distribution
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot( x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
# This is a normal distribution
```

OBSERVTATIONS
    1. Not rated count is very high
    2. Maximum number of rating are between 2.5 to 3.4 --> Seemingly looks like a normal distribution


```python
## Count Plot
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
